```python
import os
import glob

import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

# Small batch, reward_to_go, normalize_advantage

```python
infiles = sorted(glob.glob('./data/sb_rtg_na/RoboschoolInvertedPendulum-v1/*.csv'), 
                 key=lambda s: int(os.path.basename(s).replace('.csv', '')))
print(len(infiles))
```

```python
fig, axes = plt.subplots(1, 3, figsize=(14, 4))
axes = axes.ravel()

for k, col in enumerate(['avg_ret', 'min_ret', 'max_ret']):
    ax = axes[k]
    
    dfs = []
    for f in infiles:
        _df = pd.read_csv(f)
        dfs.append(_df[col])

    cat_df = pd.concat(dfs, axis=1)

    avg = cat_df.mean(axis=1)
    std = cat_df.std(axis=1)

    xs = cat_df.index.values + 1

    ax.plot(xs, avg)
    ax.fill_between(xs, avg - std, avg+std, alpha=0.2)
    ax.set_title(col)
    ax.grid()
```
