Here I report that DAgger performs better than behavioral cloning on the task **Walker2d-v1**.

In DAgger,

1. each training session goes through the training data for N epochs, 
1. each data aggregation session take 100 time steps
1. repeat for M DAgger rounds.

So in total, the training takes NM epochs.

Similarly, to make the comparison fair, regular training also takes NM epochs
